# **Loading the surprise data and its showing and converting them into Json format**

*You can replace the data_path and data_label_path with the path of your surpise.data and suprise.solution files inside the surprise data.*

```python
import json
import pandas as pd
data_path="/content/drive/MyDrive/intent_classification_It_bombay/surprise_data/surprise.data"
data_label_path="/content/drive/MyDrive/intent_classification_It_bombay/surprise_data/surprise.solution"
with open(data_path,'r') as file:
	files=file.readlines()
# print(files)
# print(len(files))
print('\n')
# print(files[0])
files_list=[]
for line in files:
	line=line.strip('\n')
	line=json.loads(line)
	files_list.append(line)
# print(dict)
print(len(files_list))
# for element in dict:
	# print(element)
	# print('\n')
```

```python
# files
```

```python
with open(data_label_path,'r') as file:
	labels=file.readlines()
print(labels)
label_list=[]
for label in labels:
  label.strip('\n')
  label=json.loads(label)
  label_list.append(label)
print(label_list)
```

```python
from transformers import AutoTokenizer
"loading the pretratined model from the hugging face by IBM trained on another intent classification task on 180 labels"
model_name="ibm/roberta-large-vira-intents" # by the IBM on 180 labels

tokenizer=AutoTokenizer.from_pretrained(model_name) #loading the tokenizer

```

# **Converting the files list from above (contain utterances for the surprise dataset) and label list(containing intent for those utterances) into the dictionary having keys as indoml id, utt and indoml id and intent**

```python
import pandas as pd
indoml_id_file=[]
utt=[]
indoml_id_label=[]
intent=[]


files_dict={}
label_dict={}
for file in files_list:
  indoml_id_file.append(file['indoml_id'])
  utt.append(file['utt'])
  files_dict['indoml_id']=indoml_id_file
  files_dict['utt']=utt

for file in label_list:
  indoml_id_label.append(file['indoml_id'])
  intent.append(file['intent'])
  label_dict['indoml_id']=indoml_id_label
  label_dict['intent']=intent


```

# **Converting files_dict and labels_dict obtained from above into  into pandas dataframe**

```python
files_pd=pd.DataFrame.from_dict(files_dict)
files_pd
```

```python
labels_pd=pd.DataFrame.from_dict(label_dict)
labels_pd
```

```python
labels_pd['intent'].unique() # we have here a total of 150 unique intents
```

# **merging both dataframe obtained above on the basis of common indoml id**

```python
Data_pd=pd.merge(files_pd, labels_pd, on='indoml_id', how='inner')
Data_pd
```

```python
Data_pd.drop('indoml_id', axis=1, inplace=True)
```

```python
Data_pd
```

# **Encoding all the labels to be able to use them further**

```python
labels_intent_list=[]
for label in label_list:
      labels_intent_list.append(label['intent'])
unique_labels_list=[]
for x in labels_intent_list:
    if x not in unique_labels_list:
        unique_labels_list.append(x)
# unique_labels_list, len(unique_labels_list)

label2id={}
id2label={}
for i, label in enumerate(unique_labels_list):
    label2id[label]=i
    id2label[i]=label
# converting the intent into id by label2id
labels=[label2id[intent] for intent in Data_pd['intent']]
labels[10], len(labels)
```

```python
Data_pd['labels']=labels
Data_pd
```

```python
from datasets import load_dataset
from sentence_transformers.losses import CosineSimilarityLoss
from setfit import SetFitModel, SetFitTrainer # importing SetFitmodel and SetFittrainer
```

```python
train_dataset=load_dataset('csv', data_files='/content/train.csv') # loading the training csv file
```

```python
test_dataset=load_dataset('csv', data_files='/content/test.csv') # loading the testinf csv file
```

```python
train_dataset['train'], test_dataset['train']
```

```python
model_name="ibm/roberta-large-vira-intents" # by the IBM on 180 labels # pretrainjed roberta large intent classification model
model_id=model_name
model=SetFitModel.from_pretrained(model_id) # loading the model
```

```python
trainer=SetFitTrainer(
    model=model,
    train_dataset=train_dataset['train'],
    eval_dataset=test_dataset['train'],
    loss_class=CosineSimilarityLoss,
    metric='accuracy',
    batch_size=32,
    num_iterations=20,  # Find the closest 5 stories of the corpus for each query sentence based on cosine similarity
    num_epochs=5,
    column_mapping={'text':'text','labels':'label'}

)
```

```python
trainer.train()
```

```python
metrics=trainer.evaluate() # prediction on the validation dataset
metrics
```
